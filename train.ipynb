```python
import os

label_dir = "weapon-detection-1/train/labels"

empty = 0
bad = 0
total = 0

for file in os.listdir(label_dir):
    path = os.path.join(label_dir, file)
    total += 1

    if os.path.getsize(path) == 0:
        empty += 1
        continue

    with open(path, "r") as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) != 5:
                bad += 1

print("Total labels:", total)
print("Empty labels:", empty)
print("Bad format labels:", bad)
```

```python
import os
import random
import cv2
import matplotlib.pyplot as plt

base = "weapon-detection-1/train"
img_dir = os.path.join(base, "images")
lbl_dir = os.path.join(base, "labels")

empty_files = [f for f in os.listdir(lbl_dir)
               if os.path.getsize(os.path.join(lbl_dir, f)) == 0]

print("Empty labels:", len(empty_files))

# show 9 random samples
sample = random.sample(empty_files, min(9, len(empty_files)))

plt.figure(figsize=(10, 10))
for i, f in enumerate(sample, 1):
    img_name = f.replace(".txt", ".jpg")
    img_path = os.path.join(img_dir, img_name)

    if os.path.exists(img_path):
        img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
        plt.subplot(3, 3, i)
        plt.imshow(img)
        plt.title("EMPTY LABEL")
        plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
non_empty = [f for f in os.listdir(lbl_dir)
             if os.path.getsize(os.path.join(lbl_dir, f)) > 0]

sample = random.sample(non_empty, min(9, len(non_empty)))

plt.figure(figsize=(10, 10))
for i, f in enumerate(sample, 1):
    img_name = f.replace(".txt", ".jpg")
    img_path = os.path.join(img_dir, img_name)

    if os.path.exists(img_path):
        img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
        plt.subplot(3, 3, i)
        plt.imshow(img)
        plt.title("HAS LABEL")
        plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import os

base = "weapon-detection-1"

splits = ["train", "valid", "test"]

for split in splits:
    img_dir = os.path.join(base, split, "images")
    lbl_dir = os.path.join(base, split, "labels")

    if not os.path.exists(img_dir):
        print(f"{split}: folder missing")
        continue

    images = [f for f in os.listdir(img_dir) if f.endswith((".jpg", ".png", ".jpeg"))]
    labels = os.listdir(lbl_dir) if os.path.exists(lbl_dir) else []

    empty = 0
    missing = 0

    # count empty labels
    for f in labels:
        path = os.path.join(lbl_dir, f)
        if os.path.getsize(path) == 0:
            empty += 1

    # check missing labels
    for img in images:
        label = img.rsplit(".", 1)[0] + ".txt"
        if not os.path.exists(os.path.join(lbl_dir, label)):
            missing += 1

    print(f"\n--- {split.upper()} ---")
    print("Images:", len(images))
    print("Labels:", len(labels))
    print("Empty labels:", empty)
    print("Missing labels:", missing)
```
